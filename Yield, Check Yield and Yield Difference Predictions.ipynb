```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
```

```python
# Read in files
#genentic = pd.read_csv('Training Data/Training_Genetic_Dataset.csv')
performance = pd.read_csv('Training Data/Training_Performance_Dataset.csv')
soil = pd.read_csv('Training Data/Training_Soil_Dataset.csv') 
weather = pd.read_csv('Training Data/Training_Weather_Dataset.csv') 
```

```python
soil.drop(['Latitude','Longitude'], axis = 1, inplace = True)
weather.drop(['Latitude','Longitude'], axis = 1, inplace = True)
perf = pd.merge(performance, soil, how='left', on='Location_ID')
perf = pd.merge(perf, weather, how='left', on=['Location_ID','Year'])
perf.head()
```

## Boosted Tree on Yield Difference

```python
# Set-up the predictors and the response  
X = perf.loc[:,'s1':'w_6_12']
Y = perf.loc[:,'Yield_Difference']

# Divide into a test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y)

# Fit a Model
from sklearn.ensemble import GradientBoostingRegressor   # 1. choose the model class
bst_tree = GradientBoostingRegressor(n_estimators=500, max_depth=6, learning_rate=0.5, loss='huber', alpha=0.95) # 2. instantiate model
bst_tree.fit(X_train, y_train)                           # 3. fit a model to the data
y_hat = bst_tree.predict(X_test)                         # 4. predict on the new data

# Evaluate Fit
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_hat)
r2 = r2_score(y_test, y_hat)

print('MSE: ',mse)
print('R-squared: ',r2)
```

## Boosted Tree on Just Yield

```python
# Set-up the predictors and the response  
X = perf.loc[:,'s1':'w_6_12']
Y = perf.loc[:,'Yield']

# Divide into a test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y)

# Fit a Model
from sklearn.ensemble import GradientBoostingRegressor   # 1. choose the model class
bst_tree_y = GradientBoostingRegressor(n_estimators=500, max_depth=6, learning_rate=0.5, loss='huber', alpha=0.95) # 2. instantiate model
bst_tree_y.fit(X_train, y_train)                           # 3. fit a model to the data
y_hat = bst_tree_y.predict(X_test)                         # 4. predict on the new data

# Evaluate Fit
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_hat)
r2 = r2_score(y_test, y_hat)

print('MSE: ',mse)
print('R-squared: ',r2)
```

## Boosted Tree on Check Yield

```python
# Set-up the predictors and the response  
X = perf.loc[:,'s1':'w_6_12']
Y = perf.loc[:,'Check_Yield']

# Divide into a test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y)

# Fit a Model
from sklearn.ensemble import GradientBoostingRegressor   # 1. choose the model class
bst_tree_cy = GradientBoostingRegressor(n_estimators=500, max_depth=6, learning_rate=0.5, loss='huber', alpha=0.95) # 2. instantiate model
bst_tree_cy.fit(X_train, y_train)                           # 3. fit a model to the data
y_hat = bst_tree_cy.predict(X_test)                         # 4. predict on the new data

# Evaluate Fit
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_hat)
r2 = r2_score(y_test, y_hat)

print('MSE: ',mse)
print('R-squared: ',r2)
```

## Using the Check Yield and Yield Predictions to Calculate Yield_Difference

```python
perf['cy_pred'] = bst_tree_cy.predict(perf.loc[:,'s1':'w_6_12']) 
perf['y_pred'] = bst_tree_y.predict(perf.loc[:,'s1':'w_6_12']) 
perf['Yield_Difference_pred'] = perf['y_pred'] - perf['cy_pred']
perf.head()
```

```python
mse = mean_squared_error(perf['Yield_Difference'], perf['Yield_Difference_pred'])
r2 = r2_score(perf['Yield_Difference'], perf['Yield_Difference_pred'])

print('MSE: ',mse)
print('R-squared: ',r2)
```
