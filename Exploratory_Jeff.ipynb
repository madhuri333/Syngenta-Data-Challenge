## Boosted Tree using Location

```python
# Prepare the data
X = perf[['Location_ID']]
Y = perf['Yield']

from sklearn.ensemble import GradientBoostingRegressor   # 1. choose the model class
bst_tree = GradientBoostingRegressor(n_estimators=500,max_depth=5,learning_rate=0.1,loss='huber',alpha=0.95)           # 2. instantiate model
bst_tree.fit(X,Y)                                        # 3. fit a model to the data
yfit = bst_tree.predict(X)                    # 4. save data to original dataframe
yfit
```

```python
mse = mean_squared_error(perf[['Yield']], yfit)
r2 = r2_score(perf[['Yield']], yfit)

print("MSE: %.4f" % mse)
print("R2: %.4f" % r2)
```
